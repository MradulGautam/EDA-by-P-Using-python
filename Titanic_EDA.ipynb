```python
survival_counts = test_t['Survived'].value_counts()

plt.pie(survival_counts, labels=['Did Not Survive', 'Survived'], autopct='%1.1f%%', startangle=90, colors=['lightcoral', 'lightgreen'])
plt.title('Survival Distribution')
plt.axis('equal')
plt.show()
```
